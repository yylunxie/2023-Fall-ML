```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = scipy.io.loadmat('11HW1_KmeanData.mat')
```

```python
ks = np.array([4, 5, 6])
f, ax = plt.subplots(3, 3, figsize=(20, 15))
ax[0][0].axis('off')
ax[0][2].axis('off')
ax[0][1].scatter(data["X"][0, :], data["X"][1, :], c='skyblue')
ax[0][1].scatter(data["Y"][0, :], data["Y"][1, :], c='lightcoral')
ax[0][1].set_title('Original Data', fontsize=15)
for i in range(3):
    kmeans = KMeans(n_clusters=ks[i], random_state=0).fit(data["X"].T)
    ax[1][i].scatter(data["X"][0, :], data["X"][1, :], c=kmeans.labels_)
    ax[1][i].set_title("dataset X, k = " + str(ks[i]), fontsize=15)
for i in range(3):
    kmeans = KMeans(n_clusters=ks[i], random_state=0).fit(data["Y"].T)
    ax[2][i].scatter(data["Y"][0, :], data["Y"][1, :], c=kmeans.labels_)
    ax[2][i].set_title("dataset Y, k = " + str(ks[i]), fontsize=15)
plt.tight_layout()
plt.savefig("11HW1_Kmean.png", dpi=300)
plt.show()
```

k = 5 got the best result
