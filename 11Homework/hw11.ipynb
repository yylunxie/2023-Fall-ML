## Problem 1

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = scipy.io.loadmat('11HW1_KmeanData.mat')
```

```python
ks = np.array([4, 5, 6])
f, ax = plt.subplots(3, 3, figsize=(20, 15))
ax[0][0].axis('off')
ax[0][2].axis('off')
ax[0][1].scatter(data["X"][0, :], data["X"][1, :], c='skyblue')
ax[0][1].scatter(data["Y"][0, :], data["Y"][1, :], c='lightcoral')
ax[0][1].set_title('Original Data', fontsize=15)
for i in range(3):
    kmeans = KMeans(n_clusters=ks[i], random_state=0).fit(data["X"].T)
    ax[1][i].scatter(data["X"][0, :], data["X"][1, :], c=kmeans.labels_)
    ax[1][i].set_title("dataset X, k = " + str(ks[i]), fontsize=15)
for i in range(3):
    kmeans = KMeans(n_clusters=ks[i], random_state=0).fit(data["Y"].T)
    ax[2][i].scatter(data["Y"][0, :], data["Y"][1, :], c=kmeans.labels_)
    ax[2][i].set_title("dataset Y, k = " + str(ks[i]), fontsize=15)
plt.tight_layout()
plt.savefig("11HW1_Kmean.png", dpi=300)
plt.show()
```

k = 5 got the best result

## Problem 2

```python
import cv2 as cv

img = cv.imread('11HW2_PatchPanels.jpg')
img_gray = cv.imread('11HW2_PatchPanelsGrey.jpeg', cv.IMREAD_GRAYSCALE)

plt.figure(figsize=(20, 12))
clust = [2, 5, 10, 15, 20]

for i in range(5):
    # print(k_gray)
    kmeans = KMeans(n_clusters=clust[i])
    kmeans.fit(img_gray.reshape(-1, 1))
    predict = kmeans.predict(img_gray.reshape(-1, 1)).astype(np.float32)
    predict /= predict.max()
    predict *= 255
    k_gray = np.reshape(predict, img_gray.shape)
    # k_gray /= k_gray.max()
    # k_gray *= 255
    plt.subplot(231+i)
    plt.title(f'K={clust[i]}')
    plt.imshow(k_gray, cmap='gray')
plt.show()


```

```python
plt.figure(figsize=(20, 12))

for i in range(5):
    # print(k_gray)
    kmeans = KMeans(n_clusters=clust[i])
    kmeans.fit(img.reshape(-1, 3))
    predict = kmeans.predict(img.reshape(-1, 3)).astype(np.float32)
    predict /= predict.max()
    predict *= 255
    k_color = np.reshape(predict, (img.shape[0], img.shape[1]))
    # k_gray /= k_gray.max()
    # k_gray *= 255
    plt.subplot(231+i)
    plt.title(f'K={clust[i]}')
    plt.imshow(k_color, cmap='gray')
plt.show()
```

Assume img is read in unint8 format. Then the compresion ratio of k is k/255.<br>

for gray img, k=5 is sutable for it, since the img keep the information as well as clean display. During high clusters, the image appear chaos.<br>

for color img k=10 retaining an appropriate amount of information.
